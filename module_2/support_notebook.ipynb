```python
data = """
Input size (n), Total Time (ms)
1000,74
2500,72
5000,75
7500,76
10000,83
25000,88
50000,92
75000,98
100000,105
150000,118
200000,127

Input size (n), Total Time (ms)
1000,137
2500,240
5000,478
7500,694
10000,903
25000,2138
50000,4245
75000,6421
100000,8639
150000,13073
200000,17546
"""
```

```python
import matplotlib.pyplot as plt
import numpy as np

def parse_data(data):
    series = []
    current_series = []
    
    for line in data.strip().split('\n'):
        line = line.strip()
        if line.startswith("Input size"):
            if current_series:
                series.append(np.array(current_series))
                current_series = []
        elif line:
            parts = line.split(',')
            current_series.append((int(parts[0]), int(parts[1])))
    
    if current_series:
        series.append(np.array(current_series))
    
    return series

# Parse input
data_series = parse_data(data)

# Plotting
plt.figure(figsize=(10, 6))
for idx, series in enumerate(data_series):
    x, y = series[:, 0], series[:, 1]
    plt.plot(x, y, marker='o', label=f'Series {idx + 1}')

plt.xlabel("Input size (n)")
plt.ylabel("Total Time (ms)")
plt.title("Performance Comparison")
plt.legend()
plt.grid(True)
plt.show()
```
